```python
import matplotlib.pyplot as plt
import numpy as np

# Data
x = np.arange(1, 11)
y = x**2
data = np.random.randn(1000)

# Create 1 row, 2 columns of subplots
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# === Left: Line chart ===
axes[0].plot(x, y, marker='o', linestyle='-.', color='m', label="Magenta dash-dot with circles")
axes[0].plot(x, y*1.5, marker='s', linestyle='--', color='g', label="Green dashed with squares")
axes[0].plot(x, y*0.5, marker='^', linestyle=':', color='r', label="Red dotted with triangles")
axes[0].plot(x, y*2, marker='*', linestyle='-', color='b', label="Blue solid with stars")
axes[0].plot(x, y*0.8, marker='D', linestyle='--', color='orange', label="Orange dashed with diamonds")
axes[0].set_title("Line Chart using Matplotlib")
axes[0].set_xlabel("X-axis")
axes[0].set_ylabel("Y-axis")
axes[0].legend()
axes[0].grid(True)

# === Right: Histogram ===
axes[1].hist(data, bins=20, color='skyblue', edgecolor='black')
axes[1].set_title("Histogram using Matplotlib")
axes[1].set_xlabel("Bins")
axes[1].set_ylabel("Frequency")
axes[1].grid(True)

# Adjust layout
plt.tight_layout()
plt.show()
```

# Seaborn

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd

df = sns.load_dataset("iris")

# Pairplot (works because hue is categorical)
sns.pairplot(df, hue="species")
plt.show()

# Heatmap (numeric columns only)
num = df.select_dtypes(include="number")     # sepal_length, sepal_width, petal_length, petal_width
corr = num.corr()                            # or: df.corr(numeric_only=True)

plt.figure(figsize=(6,4))
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap (Numeric Features)")
plt.tight_layout()
plt.show()
```

# Plotly

```python
import plotly.express as px

# Scatter plot
fig = px.scatter(df, x="sepal_length", y="petal_width", color="species",
                 title="Interactive Scatter Plot using Plotly")
fig.write_html("scatter.html", auto_open=True)

# 3D Scatter
fig = px.scatter_3d(df, x='sepal_length', y='sepal_width',
                    z='petal_length', color='species')
fig.write_html("scatter1.html", auto_open=True)
```
