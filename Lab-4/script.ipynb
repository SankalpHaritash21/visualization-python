```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset("tips")
```

```python
tips.head()
```

```python
# Filter: only dinner time and bills greater than 20

filter_data= tips[(tips['time']=='Dinner') & (tips['total_bill']>20)]
sns.scatterplot(data=filter_data, x="total_bill", y="tip", hue="sex")
plt.title("Scatter plot of Tips vs Total Bill (Dinner, Bill > 20)")
plt.show()
```

```python
# Visualization 2: Histogram of tip amounts (filtered data)
sns.histplot(filter_data['tip'], bins=10, kde=True)
plt.title("Distribution of Tips (Dinner, Bill > 20)")
plt.show()
```

```python
# Visualization 3: Bar plot by day (filtered data)
sns.barplot(data=filter_data, x="day", y="tip", estimator="mean")
plt.title("Average Tip by Day (Dinner, Bill > 20)")
plt.show()
```

```python
# Filter only male customers who tipped more than 5

data1= tips[(tips['sex']=='Male') & (tips['tip']>5)]

sns.scatterplot(data=data1, x="total_bill", y="tip", hue="day")
plt.title("Scatter plot of Tips vs Total Bill (Male, Tip > 5)")
plt.show()
```

```python
sns.barplot(data=data1, x="day", y="tip", estimator="mean")
plt.title("Average Tip by Day (Male, Tip > 5)")
plt.show()
```
